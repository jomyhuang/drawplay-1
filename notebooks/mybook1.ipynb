```python
# 导入常用库
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 创建示例数据
np.random.seed(42)
data = {
    '年龄': np.random.normal(25, 5, 100),
    '收入': np.random.normal(8000, 2000, 100),
    '消费': np.random.normal(5000, 1500, 100),
    '类别': np.random.choice(['A', 'B', 'C'], 100)
}

# 创建DataFrame
df = pd.DataFrame(data)

# 基础统计分析
print("基础统计信息：")
print(df.describe())

# 创建可视化图表
plt.figure(figsize=(15, 5))

# 子图1：年龄分布直方图
plt.subplot(131)
sns.histplot(data=df, x='年龄')
plt.title('年龄分布')

# 子图2：收入和消费的散点图
plt.subplot(132)
plt.scatter(df['收入'], df['消费'])
plt.xlabel('收入')
plt.ylabel('消费')
plt.title('收入vs消费')

# 子图3：不同类别的收入箱线图
plt.subplot(133)
sns.boxplot(data=df, x='类别', y='收入')
plt.title('各类别收入分布')

plt.tight_layout()
plt.show()

# 相关性分析
print("\n相关性矩阵：")
print(df[['年龄', '收入', '消费']].corr())
```
